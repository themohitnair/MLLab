## Design an experiment to investigate the impact of varying the number of trees in a Random Forest classifier on its performance for a given dataset. Write Python code to implement the Random Forest algorithm with different numbers of trees and evaluate its classification performance using appropriate evaluation metrics.

```python
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()

X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.25)
```

```python
no_of_trees_list = list(range(2, 22, 2))

accuracies = []
precisions = []
recalls = []
roc_auc_scores = []
f1_scores = []
```

```python
for no_of_trees in no_of_trees_list:
    model = RandomForestClassifier(n_estimators=no_of_trees)

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_probs = model.predict_proba(X_test)[:, 1]

    accuracies.append(accuracy_score(y_pred=y_pred, y_true=y_test))
    precisions.append(precision_score(y_pred=y_pred, y_true=y_test))
    recalls.append(recall_score(y_pred=y_pred, y_true=y_test))
    roc_auc_scores.append(roc_auc_score(y_score=y_probs, y_true=y_test))
    f1_scores.append(f1_score(y_pred=y_pred, y_true=y_test))  
    
```

```python
plt.figure(figsize=(8, 6))
plt.grid()
plt.plot(no_of_trees_list, accuracies, color="red", marker="o")
plt.xlabel("Number of trees")
plt.ylabel("Accuracy")
plt.title("Accuracy-Trees curve")
```

```python
plt.figure(figsize=(8, 6))
plt.grid()
plt.plot(no_of_trees_list, precisions, color="blue", marker="o")
plt.xlabel("Number of trees")
plt.ylabel("Precision")
plt.title("Precision-Trees curve")
```

```python
plt.figure(figsize=(8, 6))
plt.grid()
plt.plot(no_of_trees_list, recalls, color="yellow", marker="o")
plt.xlabel("Number of trees")
plt.ylabel("Recall")
plt.title("Recall-Trees curve")
```

```python
plt.figure(figsize=(8, 6))
plt.grid()
plt.plot(no_of_trees_list, f1_scores, color="green", marker="o")
plt.xlabel("Number of trees")
plt.ylabel("F1 Score")
plt.title("F1-Trees curve")
```

```python
plt.figure(figsize=(8, 6))
plt.grid()
plt.plot(no_of_trees_list, roc_auc_scores, color="red", marker="o")
plt.xlabel("Number of trees")
plt.ylabel("AUC")
plt.title("AUC-Trees curve")
```
