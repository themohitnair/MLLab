```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
import math
```

```python
df = pd.read_csv("advertising.csv")
df = df.drop(columns=["Unnamed: 0"])
```

```python
df.head()
```

```python
X = df.drop(columns=["sales"])
y = df["sales"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
def plot_data(adv_type: str, color: str):
    plt.figure(figsize=(10, 8))
    plt.scatter(X[adv_type], y, color=color)
    plt.xlabel(f"Expenditure on {adv_type} advertisements")
    plt.ylabel("Sales")
    plt.title(f"Sales vs {adv_type} expenditure")
    plt.grid()
    plt.show()

def train_and_plot_regression_line(adv_type: str, color: str):
    model = LinearRegression()
    model.fit(X_train[[adv_type]], y_train)

    y_pred = model.predict(X_test[[adv_type]])

    rmse = math
    
    plt.figure(figsize=(10, 8))
    plt.plot(X_test[adv_type], y_pred, color="black", linestyle=":", label="Regression Line (Predicted)")
    plt.scatter(X_test[adv_type], y_test, color=color, label="Test Data")
    plt.xlabel(f"Expenditure on {adv_type} advertisements")
    plt.ylabel("Sales")
    plt.title(f"Sales vs {adv_type} expenditure")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_data(adv_type="radio", color="red")
plot_data(adv_type="TV", color="navy")
plot_data(adv_type="newspaper", color="green")
```

```python
train_and_plot_regression_line("radio", "red")
train_and_plot_regression_line("TV", "navy")
train_and_plot_regression_line("newspaper", "green")
```
