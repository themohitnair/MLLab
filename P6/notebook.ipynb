```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
```

```python
df = pd.read_csv("play_tennis.csv")
```

```python
df.head()
```

```python
outlook_le = LabelEncoder()
temp_le = LabelEncoder()
hum_le = LabelEncoder()
wind_le = LabelEncoder()
play_le = LabelEncoder()
```

```python
df["outlook_enc"] = outlook_le.fit_transform(df["Outlook"])
df["temp_enc"] = temp_le.fit_transform(df["Temperature"])
df["hum_enc"] = hum_le.fit_transform(df["Humidity"])
df["wind_enc"] = wind_le.fit_transform(df["Wind"])
df["play_enc"] = play_le.fit_transform(df["Play Tennis"])
```

```python
X = df[["outlook_enc", "temp_enc", "hum_enc", "wind_enc"]]
y = df["play_enc"]
```

```python
model = CategoricalNB()
model.fit(X, y)
```

```python
test_df = pd.DataFrame({
    "Outlook": ["Sunny", "Overcast"],
    "Temperature": ["Hot", "Cool"],
    "Humidity": ["High", "Normal"],
    "Wind": ["Yes", "No"]
})
```

```python
test_df["outlook_enc"] = outlook_le.fit_transform(test_df["Outlook"])
test_df["temp_enc"] = temp_le.fit_transform(test_df["Temperature"])
test_df["hum_enc"] = hum_le.fit_transform(test_df["Humidity"])
test_df["wind_enc"] = wind_le.fit_transform(test_df["Wind"])

X_synthetic = test_df[["outlook_enc", "temp_enc", "hum_enc", "wind_enc"]]
```

```python
predictions = model.predict(X_synthetic)

prediction_probabilities = model.predict_proba(X_synthetic)
```

```python
predicted_labels = play_le.inverse_transform(predictions)
```

```python
for i in range(len(test_df)):
    print(f"Input: {dict(test_df.iloc[i])}")
    print(f"Output: {predicted_labels[i]}")
    print(f"Confidence: No -> {prediction_probabilities[i][0]}, Yes -> {prediction_probabilities[i][1]}")
```
